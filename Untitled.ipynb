```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image, display

def generate_and_display_latency_chart(file_path):
    """
    Loads data, calculates latency deviation, saves a static PNG,
    and then explicitly displays that PNG.
    """
    try:
        # Load the dataset
        df = pd.read_csv(file_path)

        # --- Data Cleaning ---
        df.columns = df.columns.str.replace('\n', ' ', regex=False).str.strip()
        metric_col = 'Total Response Time (s)'
        
        if metric_col not in df.columns:
            print(f"Error: Column '{metric_col}' not found.")
            return

        df['Model_Provider'] = df['API Provider'] + ' - ' + df['Model']
        df[metric_col] = pd.to_numeric(df[metric_col], errors='coerce')
        df_cleaned = df.dropna(subset=[metric_col])

        if df_cleaned.empty:
            print(f"No valid numerical data found in '{metric_col}'.")
            return

        # --- Deviation Analysis ---
        average_latency = df_cleaned[metric_col].mean()
        df_cleaned['Deviation (s)'] = df_cleaned[metric_col] - average_latency
        df_cleaned = df_cleaned.sort_values(by='Deviation (s)')

        # --- Visualization with Matplotlib ---
        print(f"Generating chart image... Average Total Response Time: {average_latency:.2f} seconds")
        colors = ['#4393c3' if x < 0 else '#d6604d' for x in df_cleaned['Deviation (s)']]
        num_models = len(df_cleaned)
        fig_height = max(8, num_models * 0.4) 

        fig, ax = plt.subplots(figsize=(12, fig_height))
        y_pos = np.arange(len(df_cleaned['Model_Provider']))
        ax.barh(y_pos, df_cleaned['Deviation (s)'], color=colors, align='center')

        ax.set_yticks(y_pos)
        ax.set_yticklabels(df_cleaned['Model_Provider'], fontsize=10)
        ax.invert_yaxis()  

        ax.axvline(0, color='black', linestyle='--', linewidth=1)
        ax.text(0, ax.get_ylim()[0] * 1.01, '  Average', verticalalignment='top', horizontalalignment='center', color='black', fontsize=10)

        ax.set_xlabel(f'Deviation from Average ({average_latency:.2f}s)', fontsize=12)
        ax.set_title('Deviation in Total Response Time from Average\n(Faster vs. Slower Models)', fontsize=16, pad=20)
        ax.xaxis.grid(True, linestyle='--', which='major', color='grey', alpha=0.25)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        plt.tight_layout(pad=1.5)

        # --- Save and Display ---
        chart_filename = "latency_deviation_chart.png"
        plt.savefig(chart_filename, dpi=150)
        plt.close(fig) # Close the plot to free memory
        
        print(f"Chart saved as '{chart_filename}'. Now displaying...")
        
        # Explicitly display the saved image
        display(Image(filename=chart_filename))

    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Run the analysis and display the image
generate_and_display_latency_chart("ApiProviders.csv")
```
